```python
# Case Study: Library Management System
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author
        self.available = True

    def borrow(self):
        if self.available:
            self.available = False
            print(f"Borrowed book: {self.title} by {self.author}")
        else:
            print(f"Book '{self.title}' is not available.")

    def return_book(self):
        if not self.available:
            self.available = True
            print(f"Returned book: {self.title} by {self.author}")
        else:
            print(f"Book '{self.title}' is already available.")


class Library:
    def __init__(self):
        self.books = []

    def add_book(self, book):
        self.books.append(book)

    def display_available_books(self):
        print("Available Books:")
        for book in self.books:
            if book.available:
                print(f"- {book.title} by {book.author}")


# Create some books
book1 = Book("The Lord of the Rings", "J.R.R. Tolkien")
book2 = Book("Pride and Prejudice", "Jane Austen")
book3 = Book("The Hitchhiker's Guide to the Galaxy", "Douglas Adams")

# Create a library
library = Library()

# Add books to the library
library.add_book(book1)
library.add_book(book2)
library.add_book(book3)

# Display available books
library.display_available_books()

print("\nBorrow and return books:")
book1.borrow()
book2.borrow()
book3.borrow()  # Should print that book is not available

print("\n")
library.display_available_books()  # Should show no books available

book2.return_book()
book3.return_book()  # Should print that book is already available

print("\n")
library.display_available_books()  # Should show book2 available
```

```python
# Experiment: Financial Portfolio Management System
class Asset:
    def __init__(self, name, quantity, price):
        self.name = name
        self.quantity = quantity
        self.price = price

    def value(self):
        return self.quantity * self.price


class Portfolio:
    def __init__(self):
        self.assets = []

    def add_asset(self, asset):
        self.assets.append(asset)
        print(f"Asset '{asset.name}' added to the portfolio.")

    def remove_asset(self, asset_name):
        self.assets = [a for a in self.assets if a.name != asset_name]
        print(f"Asset '{asset_name}' removed from the portfolio.")

    def total_value(self):
        return sum(asset.value() for asset in self.assets)

    def display_assets(self):
        if not self.assets:
            print("No assets in the portfolio.")
        else:
            print("Portfolio assets:")
            for asset in self.assets:
                print(f"- {asset.name}: {asset.quantity} units at Rs.{asset.price:.2f} each")


# Example usage:
if __name__ == "__main__":
    my_portfolio = Portfolio()

    # Adding assets
    stock1 = Asset("AAPL", 10, 150.0)
    stock2 = Asset("GOOG", 5, 2800.0)
    cash = Asset("Cash", 5000, 1.0)
    my_portfolio.add_asset(stock1)
    my_portfolio.add_asset(stock2)
    my_portfolio.add_asset(cash)

    # Display assets
    print("\n")
    my_portfolio.display_assets()

    # Calculate total portfolio value
    total_value = my_portfolio.total_value()
    print(f"Total portfolio value: Rs.{total_value:.2f}")
```
