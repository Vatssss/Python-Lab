```python
'''
Experiment-3: Write a program to create a set of user-defined functions to perform basic
mathematical operations (addition, subtraction, multiplication, and division) and explore different
ways of passing parameters to functions.
'''

def add(x,y):
    return x+y

def subtract(x,y):
    return x-y

def multiply(x,y):
    return x*y

def divide(x,y):
    return x/y

x = int(input("Enter value of x:"))
y = int(input("Enter value of y:"))

print("Positional arguments:")
print(add(x,y))
print(subtract(x,y))
print(multiply(x,y))
print(divide(x,y))

print("Keyword arguments:")
print(add(y=8,x=9))
print(subtract(y=8,x=9))
print(multiply(y=8,x=9))
print(divide(y=8,x=9))

print("Variable-length arguments:")
def sum_all(*numbers):  
    total = 0
    for number in numbers:
        total += number
    return total

print(sum_all(1, 2, 3, 4, 5))
```

```python
'''
Experiment-4: Create simple lambda functions for basic operations like addition, subtraction,
multiplication, and division.
 Use lambda functions with built-in functions like filter() to filter elements from a list.
 Use lambda functions with built-in functions like map() to perform operations on each element
of a list.
 Use lambda functions with the sorted() function to customize sorting.
'''

def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    if y == 0:
        raise ZeroDivisionError("Cannot divide by zero")
    return x / y

# Lambda functions for basic operations
add_lambda = lambda x, y: x + y
subtract_lambda = lambda x, y: x - y
multiply_lambda = lambda x, y: x * y
divide_lambda = lambda x, y: x / y if y != 0 else None  # Handle division by zero

# Sample data
numbers = [1, 2, 3, 4, 5, 6]
names = ["Alice", "bob", "Charlie", "david"]

# Filter even numbers using lambda with filter()
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print("Even numbers using filter and lambda:", even_numbers)

# Square each number using lambda with map()
squared_numbers = list(map(lambda x: x**2, numbers))
print("Squared numbers using map and lambda:", squared_numbers)

# Sort names case-insensitively using lambda with sorted()
sorted_names = sorted(names, key=lambda name: name.lower())
print("Sorted names (case-insensitive) using sorted and lambda:", sorted_names)

# Example usage of other functions
result_add = add(3, 5)
result_subtract = subtract(10, 4)
result_multiply = multiply(2, 6)
result_divide = divide(12, 3)

print("Addition:", result_add)
print("Subtraction:", result_subtract)
print("Multiplication:", result_multiply)
print("Division:", result_divide)

# Example usage of lambda functions
result_add_lambda = add_lambda(7, 8)
result_subtract_lambda = subtract_lambda(15, 9)
result_multiply_lambda = multiply_lambda(4, 3)
result_divide_lambda = divide_lambda(20, 5)

print("Addition (lambda):", result_add_lambda)
print("Subtraction (lambda):", result_subtract_lambda)
print("Multiplication (lambda):", result_multiply_lambda)
print("Division (lambda):", result_divide_lambda)
```

```python
'''
Experiment-6: Define a base class, e.g., Vehicle, with attributes such as make, model, and year.
 Implement a constructor to initialize these attributes.
 Create instances of the Vehicle class, representing different vehicles with distinct attributes.
 Display the information for each vehicle using appropriate methods.
 Overload the + operator to combine the make and model attributes of two instances.
 Display the result of this operator overloading operation.
 Create a derived class, e.g., Car, that inherits from the Vehicle class.
 Add specific attributes to the Car class, such as num_doors and fuel_type.
 Instantiate objects of both the Vehicle and Car classes.
 Display the information for each object, showcasing the inheritance structure.
'''

class Vehicle:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    def display_info(self):
        print(f"Make: {self.make}\nModel: {self.model}\nYear: {self.year}")

    def __add__(self, other):
        combined_make_model = f"{self.make} {self.model} + {other.make} {other.model}"
        return combined_make_model

class Car(Vehicle):
    def __init__(self, make, model, year, num_doors, fuel_type):
        super().__init__(make, model, year)
        self.num_doors = num_doors
        self.fuel_type = fuel_type

    def display_info(self):
        super().display_info()
        print(f"Number of doors: {self.num_doors}\nFuel type: {self.fuel_type}")

# Create Vehicle instances
vehicle1 = Vehicle("Toyota", "Camry", 2023)
vehicle2 = Vehicle("Honda", "Accord", 2024)

# Display vehicle information
vehicle1.display_info()
vehicle2.display_info()

# Overload + operator
combined_vehicles = vehicle1 + vehicle2
print("\nCombined make and model:", combined_vehicles)

# Create Car instance
car1 = Car("Ford", "Mustang", 2022, 2, "Gasoline")

# Display car information (showcasing inheritance)
car1.display_info()
```
